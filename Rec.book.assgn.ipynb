```python
user_df = df.pivot_table(index='userid',
                         columns='booktitle',
                         values='bookrating').reset_index(drop=True)
```

```python
user_df
```

```python
user_df.index = df.userid.unique()
user_df
```

```python
user_df.fillna(0, inplace=True)
user_df
```

```python
user_cossim = 1 - pairwise_distances( user_df.values,metric='cosine')
user_cossim
```

```python
user_cossim_df = pd.DataFrame(user_cossim)
```

```python
user_cossim_df.index = df.userid.unique()
user_cossim_df.columns = df.userid.unique()
```
